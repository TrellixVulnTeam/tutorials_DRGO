# Tutorial 1: Exploratory Data Analysis using Sweetviz

In this notebook, we will show how to perform basic exploratory data analysis (EDA) using open-source library sweetviz. 

## Steps 

- Download sample credit card data used in this [tutorial](https://machinelearningmastery.com/imbalanced-classification-with-the-fraudulent-credit-card-transactions-dataset/)
```
wget https://raw.githubusercontent.com/jbrownlee/Datasets/master/creditcard.csv.zip
```
- Create a pandas dataframe using the csv data
- Analyze the data using sweetviz
- Display the results

````python
from pandas import read_csv
from matplotlib import pyplot
import sweetviz as sv
````

````python
# define the dataset location
filename = 'creditcard.csv'
# load the csv file as a data frame
df = read_csv(filename, header=None)
# drop the target variable
df = df.drop(30, axis=1)
df
````

````python
report = sv.analyze(df)
report.show_notebook()
````
